### The Challenge

You need to write a program to count the number of words in a line of text. A word is any sequence of lowercase characters. In this challenge, 'Hello' is a word and 'bbaabbb' is also a word.


### Input

The input consists of a single line of text with lowercase characters and spaces. Each pair of words has exactly one space between them, and there are no spaces before or after the first word. The line can be up to 80 characters long.

### Output

Output the number of words in the input line.



```python
#validate input is 80 characters or less
while True:
  check_string = input("Please enter your sentence:\n")
  #Python's len() function returns the number of characters in a string
  if len(check_string) < 81:
    break
  else:
    print("Your sentence is too long!\nPlease ensure the sentence is no longer than 80 characters.")


#initialise num_words outside of the loop
num_words = 0

#use a for loop to iterate through the string, incrementing num_words every time we encounter a space
#given the sentence rules stated above, the number of spaces in the string will always equal the number of words minus 1
for letter in check_string:
  #the ord() function converts a character to its ASCII value
  #32 is the ASCII value of space
  if ord(letter) == 32:
    num_words = num_words + 1

#increment num_words outside loop to account for the final word
num_words = num_words + 1

print("The number of words in the sentence you entered is:",num_words)
```
